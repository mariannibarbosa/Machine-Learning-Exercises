```python
# Create an array named as 'dataset' comprised of transactions
dataset1 = [
['Bread', 'Milk'],
['Beer', 'Bread', 'Diaper', 'Eggs'],
['Beer', 'Coke', 'Diaper','Milk'],
['Beer', 'Bread', 'Diaper', 'Milk'],
['Bread', 'Coke', 'Diaper', 'Milk']
]
```

```python
# Display the records
print(dataset1)
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

# Create an object te by calling a method TransactionEncoder()
te = TransactionEncoder()

# Call fit() method to train the model
te_array1 = te.fit(dataset1).transform(dataset1)

# Transform te_array into dataframe
df1 = pd.DataFrame(te_array1, columns = te.columns_)
```

```python
df1.head()
```

```python
from mlxtend.frequent_patterns import apriori

# Calculate the sfrequenct itemsets by calling a method apriori
frequent_itemsets_ap = apriori(df1, min_support = 0.01, use_colnames = True)
```

```python
print(frequent_itemsets_ap)
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets_ap, metric = 'confidence',
                          min_threshold = 0.0001)
rules
```

```python
rules.to_csv("lecture_dataset.csv")
```
